```python
import pandas as pd
df = pd.read_csv('whatsapp_chat_data.csv')
df.head()
```

```python

counts = df['Name'].value_counts()
print(counts)
talkative = counts.idxmax()
less_talkative = counts.idxmin()
print("Most talkative:", talkative, counts.max())
print("Least talkative:", less_talkative, counts.min())
    
```

```python

# Combine Date + Time into a single datetime column
df['DateTime'] = pd.to_datetime(df['Date'].str.strip() + " " + df['Time'].str.strip(),
                                format="%d/%m/%y %I:%M %p")

# Extract hour and day
df['Hour'] = df['DateTime'].dt.hour
df['Day'] = df['DateTime'].dt.date
df['DayName'] = df['DateTime'].dt.day_name()

# Calculate most active hour
most_active_hour = df['Hour'].value_counts().idxmax()
most_active_hour_count = df['Hour'].value_counts().max()

# Calculate most active day
most_active_day = df['Day'].value_counts().idxmax()
most_active_day_count = df['Day'].value_counts().max()

print("Most Active Hour:", most_active_hour, "with", most_active_hour_count, "messages")
print("Most Active Day:", most_active_day, "with", most_active_day_count, "messages")

# Optional: frequency tables
print("\nMessage count by hour:\n", df['Hour'].value_counts())
print("\nMessage count by date:\n", df['Day'].value_counts())
```

```python
media_count = (df["Text"].str.contains("media omitted", case=False, na=False)).sum()
print("Total media messages:", media_count)
```

```python
df.columns = df.columns.str.strip()  # remove extra spaces in column names

# Robust pattern for system messages inside <>
system_messages = [
    "media omitted",
    "missed voice call",
    "missed call",
    "call declined",
    "video call declined"
]

# Build regex pattern dynamically
import re
pattern = r"<\s*(" + "|".join(system_messages) + r")\s*>"

# Flag system messages
df['IsSystemMessage'] = df['Text'].str.contains(pattern, flags=re.IGNORECASE, na=False)

# Separate flags for media and missed calls
df['IsMedia'] = df['Text'].str.contains(r"<\s*media omitted\s*>", flags=re.IGNORECASE, na=False)
df['IsMissedCall'] = df['Text'].str.contains(r"<\s*(missed voice call|missed call|call declined|video call declined)\s*>", flags=re.IGNORECASE, na=False)

# Counts
print("Total system messages:", df['IsSystemMessage'].sum())
print("Total media messages:", df['IsMedia'].sum())
print("Total missed calls:", df['IsMissedCall'].sum())
```
